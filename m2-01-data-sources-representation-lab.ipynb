```python
Dataset_A=[
    {"username": "ali01", "email": "ali@example.com", "age": 25, "signup_date": "2023-01-10"},
    {"username": "leyla_h", "email": "leyla@test.com", "age": 30, "signup_date": "2023-02-15"},
    {"username": "murad99", "email": "murad_at_gmail.com", "age": 22, "signup_date": "2023-03-01"},  
    {"username": "fidan_v", "email": "fidan@site.com", "age": "28", "signup_date": "2023-03-05"},   
    {"username": "", "email": "anonim@mail.com", "age": 40, "signup_date": "2023-03-10"},           
    {"username": "nurana19", "email": "nuranali19@mail.az", "age": 19, "signup_date": "2023-03-12"},
    {"username": "aygun_z", "email": "aygun@web.com", "age": 35, "signup_date": "2023-03-15"},
    {"username": "kamran_o", "email": "kamran@code.edu", "age": 27, "signup_date": "2023-03-20"}
]

def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []

    for entry in entries:
        username = entry.get("username")
        email = entry.get("email")
        age = entry.get("age")

        is_valid = True

        # Username check
        if not username:
            is_valid = False

        # Email check
        if not isinstance(email, str) or "@" not in email:
            is_valid = False

        # Age check
        try:
            age_int = int(age)
            if age_int <= 0:
                is_valid = False
        except:
            is_valid = False

        if is_valid:
            valid_entries.append(entry)
        else:
            invalid_entries.append(entry)

    return valid_entries, invalid_entries


valid, invalid = validate_registrations(Dataset_A)

print("Valid entries count:", len(valid))
print("Invalid entries count:", len(invalid))
```

This dataset represents user input data. Since users enter the information themselves, it is likely to contain errors such as invalid emails, missing usernames, or incorrect data types (for example, age written as text). 

In production, common quality issues would be missing values, wrong formats, duplicate accounts, or unrealistic ages. 

Before using this data in an ML pipeline, I would validate that:
- username is not empty
- email contains "@" and follows a basic format
- age is a positive integer
- signup_date has a valid date format

# Task 2: Work with Multiple Data Formats

```python
#step 1
import pandas as pd
import numpy as np
import csv
import io

np.random.seed(19)  

rides = pd.DataFrame({
    "ride_id": range(1, 501),
    "driver_id": np.random.randint(1000, 1051, 500),
    "distance_km": np.random.uniform(1.0, 30.0, 500),
    "fare_usd": np.random.uniform(5.0, 75.0, 500),
    "ride_type": np.random.choice(["standard", "premium", "pool"], 500),
    "city": np.random.choice(["Berlin", "Seoul", "Nairobi", "Toronto", "Lima"], 500),
    "timestamp": pd.date_range(start="2025-01-01", periods=500, freq="min")
})
rides.head()
```

```python
#step 2

#saving data in csv
rides.to_csv("rides.csv", index=False)
#saving data in json
rides.to_json("rides.json", orient="records", lines=True)
#saving data in parquet
rides.to_parquet("rides.parquet", index=False)
```

```python
#step 3

import os

csv_size = os.path.getsize("rides.csv")
json_size = os.path.getsize("rides.json")
parquet_size = os.path.getsize("rides.parquet")

csv_kb = csv_size / 1024
json_kb = json_size / 1024
parquet_kb = parquet_size / 1024

print(f"CSV size: {csv_kb:.2f} KB")
print(f"JSON size: {json_kb:.2f} KB")
print(f"Parquet size: {parquet_kb:.2f} KB")

ratio = csv_size / parquet_size
print(f"\nParquet is {ratio:.2f} times smaller than CSV")
```

```python
#step 4

rides_csv = pd.read_csv("rides.csv")
rides_json = pd.read_json("rides.json",orient="records",lines=True)
rides_parquet = pd.read_parquet("rides.parquet")

#comparing shapes
print("Original shape:", rides.shape)
print("CSV shape:", rides_csv.shape)
print("JSON shape:", rides_json.shape)
print("Parquet shape:", rides_parquet.shape)

#comparing data types
print("\nOriginal dtypes:")
print(rides.dtypes)
print("\nCSV dtypes:")
print(rides_csv.dtypes)
print("\nJSON dtypes:")
print(rides_json.dtypes)
print("\nParquet dtypes:")
print(rides_parquet.dtypes)

#comparing samples
print("\nOriginal sample:")
print(rides.head(3))

print("\nCSV sample:")
print(rides_csv.head(3))

print("\nJSON sample:")
print(rides_json.head(3))

print("\nParquet sample:")
print(rides_parquet.head(3))
```

All three formats kept the same shape (500 rows, 7 columns).
CSV changed the timestamp column to string because it stores data as plain text.

JSON preserved the datetime type, but the precision changed from microseconds to milliseconds.

Parquet preserved all original data types exactly, including datetime and integer precision.

Overall, Parquet kept the data most accurately.

#step 5
From the comparison, Parquet preserved the data types most faithfully. It kept the datetime column and integer precision exactly the same as in the original DataFrame. CSV changed the timestamp to string, and JSON slightly reduced the datetime precision.

Parquet was also the smallest file. This is because it uses a binary columnar format and compression, while CSV and JSON store everything as plain text.

I would choose CSV if I need a simple and human-readable format that works everywhere. I would choose JSON when working with APIs or semi-structured data. However, for analytics or machine learning tasks, I would choose Parquet because it preserves types better and is more efficient for large datasets.

# Task 3: Row-Major vs Column-Major Access Patterns

```python
#step 1
import time

np.random.seed(19)
data = np.random.rand(200_000, 20)
columns = [f"feature_{i}" for i in range(20)]
df = pd.DataFrame(data, columns=columns)

print(df.shape)

#step 2
%time df["feature_0"].mean()
```

```python
%%timeit
total = 0
for i in range(10_000):
    total += df.iloc[i]["feature_0"]
```

Column access using df["feature_0"].mean() was very fast (around 1.8 ms).

Row iteration over the first 10,000 rows using df.iloc[i] took around 90 ms.

This shows that column-based vectorized operations in pandas are much faster than manually iterating over rows. The row loop is significantly slower because it runs in Python and creates additional overhead for each access.

```python
#step 3
arr = df.to_numpy()
%time arr[:, 0].mean()
```

```python
%%timeit

total = 0
for i in range(10_000):
    total += arr[i, 0]
```

NumPy column access (arr[:, 0].mean()) was very fast, around 0.5 ms.

Row iteration over the first 10,000 rows took around 2.5 ms.

This shows that in NumPy, vectorized column operations are faster than manual row loops. However, NumPy row access is still more efficient than pandas row iteration because it works directly with contiguous memory and has less overhead.

#step 4

1) Column access in pandas is faster because pandas is mainly optimized for column-based operations. When we use df["feature_0"].mean(), the calculation runs in a vectorized way and is optimized in the background. But when we iterate row by row using df.iloc[i], it works at the Python level and creates extra overhead, so it becomes slower.

2) When using NumPy, the gap becomes smaller because NumPy works directly with arrays and does not create extra objects like pandas does. It accesses memory more directly, so even row loops are relatively faster compared to pandas.

3) This shows that both pandas and NumPy are designed mainly for column-oriented data access. The data is stored column by column in memory, so column operations are more efficient. Row-by-row processing does not match this structure, so it is slower.

```python
# Step 2 
publishers = df_books[["publisher_name", "publisher_country"]].drop_duplicates().reset_index(drop=True)
publishers["publisher_id"] = publishers.index + 1
publishers = publishers[["publisher_id","publisher_name","publisher_country"]]

books = df_books.merge(publishers, on=["publisher_name","publisher_country"])
books = books[["title","author","format","publisher_id","price"]]

joined = books.merge(publishers, on="publisher_id")

joined = joined[["title","author","format","publisher_name","publisher_country","price"]]

flat_sorted = df_books.sort_values(by=["title","format"]).reset_index(drop=True)
joined_sorted = joined.sort_values(by=["title","format"]).reset_index(drop=True)

print("Joined equals flat version:", joined_sorted.equals(flat_sorted))
```

```python
documents = []

for title, g in df_books.groupby("title"):
    documents.append({
        "title": title,
        "author": g["author"].iloc[0],
        "publisher": {
            "publisher_name": g["publisher_name"].iloc[0],
            "publisher_country": g["publisher_country"].iloc[0]
        },
        "editions": g[["format", "price"]].to_dict("records")
    })

documents[0]
```

```python
author_name = "Robert C. Martin"
publisher_to_update = "O'Reilly Media"
new_country = "Ireland"
```

```python
#flat
# Q1
flat_by_author = df_books[df_books["author"] == author_name][["title", "format", "price"]]
print("FLAT - books by author:\n", flat_by_author, "\n")

# Q2
flat_avg_price = df_books["price"].mean()
print("FLAT - average price:", flat_avg_price, "\n")

# Q3
flat_mask = df_books["publisher_name"] == publisher_to_update
flat_rows_to_change = flat_mask.sum()

df_books_flat_updated = df_books.copy()
df_books_flat_updated.loc[flat_mask, "publisher_country"] = new_country

print("FLAT - rows to change:", flat_rows_to_change)
```

```python
#normalized 

# Q1
norm_by_author = books[books["author"] == author_name][["title", "format", "price", "publisher_id"]]
print("\nNORMALIZED - books by author:\n", norm_by_author, "\n")

# Q2
norm_avg_price = books["price"].mean()
print("NORMALIZED - average price:", norm_avg_price, "\n")

# Q3
pub_mask = publishers["publisher_name"] == publisher_to_update
publishers_rows_to_change = pub_mask.sum()   

publishers_updated = publishers.copy()
publishers_updated.loc[pub_mask, "publisher_country"] = new_country

print("NORMALIZED - rows to change in publishers table:", publishers_rows_to_change)
```

```python
#Document

# Q1
doc_by_author = [d for d in documents if d["author"] == author_name]
print("\nDOCUMENT - books by author:")
for d in doc_by_author:
    print("-", d["title"], "editions:", len(d["editions"]))

# Q2
all_prices = [ed["price"] for d in documents for ed in d["editions"]]
doc_avg_price = sum(all_prices) / len(all_prices)
print("\nDOCUMENT - average price:", doc_avg_price)

# Q3
docs_to_change = sum(1 for d in documents if d["publisher"]["publisher_name"] == publisher_to_update)

documents_updated = []
for d in documents:
    d2 = d.copy()
    d2["publisher"] = d["publisher"].copy()
    if d2["publisher"]["publisher_name"] == publisher_to_update:
        d2["publisher"]["publisher_country"] = new_country
    documents_updated.append(d2)

print("\nDOCUMENT - documents to change:", docs_to_change)
```

For finding all books by a specific author, the flat and normalized versions were easier because we can directly filter by the author column. In the document version we need to loop through the documents.

For calculating the average price across all editions, the flat and normalized versions were also simpler since price is just a column and we can use `.mean()`. In the document version we first need to collect prices from the editions list.

Updating a publisher’s country is easiest in the normalized version. In this structure the publisher information is stored only once in the publishers table, so usually only one record needs to be updated. In the flat and document versions the publisher data is repeated, so multiple rows or documents may need to change.

If publisher information changes frequently, the normalized structure is better because it avoids duplicated data. If each book is usually accessed as a single record, the document-oriented version is more convenient because all related information (publisher and editions) is stored together in one document.
